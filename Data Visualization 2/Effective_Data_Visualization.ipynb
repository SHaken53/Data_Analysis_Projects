```python
# Calculate mean and standard deviation
btc_mu = round(daily_return_df['BTC'].mean(), 2)
btc_sigma = round(daily_return_df['BTC'].std(), 2)

eth_mu = round(daily_return_df['ETH'].mean(), 2)
eth_sigma = round(daily_return_df['ETH'].std(), 2)

ada_mu = round(daily_return_df['ADA'].mean(), 2)
ada_sigma = round(daily_return_df['ADA'].std(), 2)

# Set the number of bins
num_bins = 40

# Plot BTC histogram
plt.figure(figsize=(8, 6))
plt.hist(daily_return_df['BTC'], bins=num_bins, edgecolor='black', color='blue', alpha=0.7)
plt.xlabel('BTC Daily Returns')
plt.ylabel('Frequency')
plt.title('BTC Daily Returns Histogram (mean={}, std={})'.format(btc_mu, btc_sigma))
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot ETH histogram
plt.figure(figsize=(8, 6))
plt.hist(daily_return_df['ETH'], bins=num_bins, edgecolor='black', color='green', alpha=0.7)
plt.xlabel('ETH Daily Returns')
plt.ylabel('Frequency')
plt.title('ETH Daily Returns Histogram (mean={}, std={})'.format(eth_mu, eth_sigma))
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot ADA histogram
plt.figure(figsize=(8, 6))
plt.hist(daily_return_df['ADA'], bins=num_bins, edgecolor='black', color='orange', alpha=0.7)
plt.xlabel('ADA Daily Returns')
plt.ylabel('Frequency')
plt.title('ADA Daily Returns Histogram (mean={}, std={})'.format(ada_mu, ada_sigma))
plt.grid(True)
plt.tight_layout()
plt.show()
```

* While BTC and ETH's returns are close to a relatively normal distribution, ADA's returns are skewed from the right.
* Since BTC has the highest frequency among these three coins, it is the most preferred investment tool in terms of trade volume and popularity.
* Since ETH and BTC are close to normal distribution, their returns are evenly distributed, but ADA has provided investors with relatively more positive returns than other coins.

```python
# import libraries 
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```python
# Check the tail of the dataframe
cancer_df.tail(5)
```

```python
# Scatter plot between mean area and mean smoothness with class labels
plt.figure(figsize=(8, 6))

sns.scatterplot(data=cancer_df, x='mean area', y='mean smoothness', hue='target', palette='Set1')

plt.xlabel('Mean Area')
plt.ylabel('Mean Smoothness')
plt.title('Scatter Plot: Mean Area vs Mean Smoothness')
plt.legend(title='Class')

sns.despine()  # Remove outer frame or border

plt.tight_layout()
plt.show()

# Print countplot to show the number of samples in each class
plt.figure(figsize=(8, 6))

sns.countplot(data=cancer_df, x='target', palette='Set1')

plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Countplot: Class Distribution')

sns.despine()  # Remove outer frame or border

plt.tight_layout()
plt.show()
```

Ten real-valued features are computed for each cell nucleus: a) radius (mean of distances from center to points on the perimeter) b) texture (standard deviation of gray-scale values) c) perimeter d) area e) smoothness (local variation in radius lengths) f) compactness (perimeter^2 / area - 1.0) g) concavity (severity of concave portions of the contour) h) concave points (number of concave portions of the contour) i) symmetry j) fractal dimension ("coastline approximation" - 1) These 10 features(a...j) are increased to 30 with feature engineering: The mean, standard error and "worst" or largest (mean of the three largest values) of these features were computed for each image, resulting in 30 features. For instance, field 3 is Mean Radius, field 13 is Radius SE, field 23 is Worst Radius.

- **Plot the scatterplot between the mean radius and mean area. Comment on the plot** 

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=cancer_df, x='mean radius', y='mean area', hue='target', palette='Set1', marker='o', edgecolor='gray')

plt.xlabel('Mean Radius')
plt.ylabel('Mean Area')
plt.title('Scatter Plot: Mean Radius vs Mean Area')

# Remove spines (borders) of the plot
sns.despine()

# Add legend
plt.legend(title='Class')

plt.tight_layout()
plt.show()
```

* Since radius is used in area calculation, there is a one-to-one relationship between mean area and mean radius.
* Since the radius is used in the area calculation, there is a one-to-one relationship between the average area and the average radius.
* Cancer cases are concentrated in mean radius < 16 and mean area < 700.

```python
# Calculate the correlation matrix
corr_matrix = cancer_df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=False, cmap='Reds', square=True, mask=np.triu(corr_matrix))

# Remove the overlapping labels
plt.tick_params(axis='x')

# Set the title and labels
plt.title('Correlation Heatmap')
plt.xlabel('Features')
plt.ylabel('Features')

# Show the heatmap
plt.show()
```

```python
stocks_prices_df = pd.read_csv('stocks_daily_prices.csv')
stocks_prices_df.head(5)
```

```python
stocks_returns_df = pd.read_csv('stocks_daily_returns.csv')
stocks_returns_df.head(5)
```

```python
# Extract the stock symbols
stock_symbols = stocks_prices_df.columns[1:]

# Set the figure size and background color
fig, ax = plt.subplots(figsize=(15, 8))
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Define line colors and styles
line_colors = ['red', 'green', 'blue', 'orange']
line_styles = ['-', '--', ':', '-.']

# Plot the daily prices for each stock
for i, symbol in enumerate(stock_symbols):
    plt.plot(stocks_prices_df['Date'], stocks_prices_df[symbol], label=symbol, color=line_colors[i], linestyle=line_styles[i])

# Set the title and labels
plt.title('Daily Stock Prices', color='white')
plt.xlabel('Date', color='white')
plt.ylabel('Price', color='white')

# Rotate and align the x-axis labels
plt.xticks(rotation=45, ha='right', color='white')

# Reduce the number of x-axis ticks
num_ticks = 10
plt.xticks(np.linspace(0, len(stocks_prices_df)-1, num_ticks, dtype=int))

# Add gridlines
plt.grid(color='gray', linestyle='dashed')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Extract the stock symbols
stock_symbols = stocks_prices_df.columns[1:]

# Set the figure size
fig, axes = plt.subplots(len(stock_symbols), 1, figsize=(12, 10), sharex=True, facecolor='black')

# Define colors for each stock
colors = ['blue', 'orange', 'green', 'red']

# Plot the daily prices for each stock on separate subplots
for i, symbol in enumerate(stock_symbols):
    ax = axes[i]
    ax.plot(stocks_prices_df['Date'], stocks_prices_df[symbol], color=colors[i])
    ax.fill_between(stocks_prices_df['Date'], stocks_prices_df[symbol], color=colors[i], alpha=0.3)
    ax.set_ylabel('Price', color='white')
    ax.set_title(symbol, color='white')

    # Rotate and align the x-axis labels
    ax.tick_params(axis='x', rotation=45, color='white')
    ax.xaxis.label.set_color('white')
    ax.yaxis.label.set_color('white')
    ax.spines['left'].set_color('white')
    ax.spines['bottom'].set_color('white')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(axis='y', colors='white')

    # Display x-axis tick labels every 5th day
    ax.xaxis.set_major_locator(plt.MaxNLocator(5))
    
    # Add gridlines
    ax.grid(color='gray', linestyle='dashed')

# Set the common x-axis label
axes[-1].set_xlabel('Date', color='white')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Extract the returns for United Airlines and P&G
ual_returns = stocks_returns_df['UAL']
pg_returns = stocks_returns_df['PG']

# Set the figure size
plt.figure(figsize=(8, 6))

# Create the histogram for United Airlines returns
plt.hist(ual_returns, bins=40, color='red', alpha=0.7, label='United Airlines')

# Create the histogram for P&G returns
plt.hist(pg_returns, bins=40, color='blue', alpha=0.7, label='P&G')

# Calculate mean and standard deviation
ual_mean = ual_returns.mean()
ual_std = ual_returns.std()
pg_mean = pg_returns.mean()
pg_std = pg_returns.std()

# Set the axis labels and title
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.title('Histogram of United Airlines and P&G Returns')

# Add a legend
plt.legend()

# Display mean and standard deviation as text on the top-right corner of the figure
plt.text(0.95, 0.82, f'UAL\nMean: {ual_mean:.4f}\nStd: {ual_std:.4f}', 
         transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', horizontalalignment='right',
         bbox=dict(facecolor='white', edgecolor='gray', alpha=0.7))
plt.text(0.95, 0.68, f'PG\nMean: {pg_mean:.4f}\nStd: {pg_std:.4f}', 
         transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', horizontalalignment='right',
         bbox=dict(facecolor='white', edgecolor='gray', alpha=0.7))

# Display the plot
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.hist(ual_returns, bins=40, color='red', alpha=0.7, label='United Airlines')
ax1.axvline(ual_mean, color='blue', linestyle='--', linewidth=2, label='Mean UAL: {:.3f}'.format(ual_mean))
ax1.axvline(ual_mean + ual_std, color='blue', linestyle=':', linewidth=2, label='Std UAL: {:.3f}'.format(ual_std))
ax1.set_xlabel('Returns')
ax1.set_ylabel('Frequency')
ax1.set_title('Histogram of United Airlines Returns')
ax1.legend()

ax2.hist(pg_returns, bins=40, color='blue', alpha=0.7, label='P&G')
ax2.axvline(pg_mean, color='red', linestyle='--', linewidth=2, label='Mean PG: {:.3f}'.format(pg_mean))
ax2.axvline(pg_mean + pg_std, color='red', linestyle=':', linewidth=2, label='Std PG: {:.3f}'.format(pg_std))
ax2.set_xlabel('Returns')
ax2.set_ylabel('Frequency')
ax2.set_title('Histogram of P&G Returns')
ax2.legend()

plt.tight_layout()
plt.show()
```

* The histogram compares the returns of United Airlines and P&G stocks.
* The red portion represents the returns of United Airlines, while the blue portion represents the returns of P&G.
* Each bar represents the number of observations within a specific return range.
* The graph shows that the distribution of United Airlines returns has a wider range compared to P&G.
* The mean and standard deviation values are indicated on the top-right corner of the graph for both stocks.
* The mean return for United Airlines is approximately 0.0004, with a standard deviation of around 0.0272.
* The mean return for P&G is approximately 0.0004, with a standard deviation of around 0.0149.
* Based on this information, we can conclude that the returns of United Airlines exhibit higher variability, while P&G shows a more stable performance.

```python
# Calculate the correlation matrix of stocks' daily returns
correlation_matrix = stocks_returns_df.corr(numeric_only=True)

# Set the figure size
plt.figure(figsize=(10, 8))

# Define a custom color palette
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Plot the heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap=cmap, center=0)

# Set the title
plt.title('Correlation Matrix of Stocks Daily Returns')

# Rotate and align the x-axis tick labels
plt.xticks(rotation=45, ha='right')

# Rotate and align the y-axis tick labels
plt.yticks(rotation=0)

# Display the plot
plt.show()
```

* The correlation coefficient between United Airlines (UAL) and P&G is a measure of the linear relationship between their daily returns.
* Correlation coefficient values range between -1 and 1, where -1 indicates a strong negative correlation, 1 indicates a strong positive correlation, and 0 indicates no correlation.
* In this case, the correlation coefficient between UAL and P&G daily returns is a positive value, indicating a positive correlation between the two stocks.
* This means that when UAL's daily returns increase, P&G's daily returns also tend to increase, and when UAL's daily returns decrease, P&G's daily returns also tend to decrease.
* However, it's important to note that correlation does not imply causation, and other factors may influence the relationship between UAL and P&G that are not captured by the correlation coefficient alone.
